```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
positive_reviews = ["Позитивна рецензія 1", "Позитивна рецензія 2"]
negative_reviews = ["Негативна рецензія 1", "Негативна рецензія 2"]

reviews = positive_reviews + negative_reviews

tokenizer = Tokenizer()
tokenizer.fit_on_texts(reviews)
total_words = len(tokenizer.word_index) + 1

input_sequences = tokenizer.texts_to_sequences(reviews)
max_sequence_length = max([len(x) for x in input_sequences])
padded_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')

X_train = padded_sequences[:2]
y_train = np.array([1, 1])  # Мітки для позитивних рецензій
X_test = padded_sequences[2:]
y_test = np.array([0, 0])  # Мітки для негативних рецензій

# Створення моделі LSTM
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(total_words, 64, input_length=max_sequence_length),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```
