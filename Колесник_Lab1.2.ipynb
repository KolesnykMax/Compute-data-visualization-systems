```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('shopping_trends.csv', encoding='cp1251')
```

Завдання №1

Оптимізація даних

```python
print("Розмір таблиці:", df.shape)
print("Пропуски:")
print(df.isnull().sum())
print("Дублікати:", df.duplicated().sum())
print("Типи стовпців:")
print(df.dtypes)
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

1. Гістограма віку (age)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Кількість клієнтів')
plt.show()
```

2. Порівняння розподілу покупок за статтею (item_purchased)

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='item_purchased', data=df, order=df['item_purchased'].value_counts().index)
plt.title('Розподіл покупок за статтею')
plt.xlabel('Кількість покупок')
plt.ylabel('Товари')
plt.show()
```

3. Залежність між віком та сумою покупок (purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.show()
```

4. Графік розподілу оцінок товарів (review_rating)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['review_rating'], bins=5, kde=True)
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Кількість товарів')
plt.show()
```

5. Категоріальна діаграма для статі (gender)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=df)
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

6. Залежність між рейтингом і сумою покупок (review_rating і purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_rating', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг')
plt.ylabel('Сума покупок')
plt.show()
```

7. Сезонність покупок (season)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=df)
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

8. Графік розподілу розмірів одягу (size)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='size', data=df)
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

9. Кількість покупок відносно типу оплати (payment_method)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='payment_method', data=df)
plt.title('Кількість покупок за типом оплати')
plt.xlabel('Тип оплати')
plt.ylabel('Кількість покупок')
plt.show()
```

10. Залежність між кількістю попередніх покупок і сумою поточної покупки (previous_purchases і purchase_amount)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='previous_purchases', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума поточної покупки')
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Данні
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

#Графік
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Графік поверхні залежності від x та y')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
